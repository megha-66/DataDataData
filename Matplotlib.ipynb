```python
arr1 = [1,2,3,4,5]
arr2 = [6,7,8,9,10]

import matplotlib.pyplot as plt
colors = ['r','b','g','c','m']
sizes = [100,200,300,400,500]

plt.scatter(arr1,arr2, c = colors, s = sizes )  #scatter plot with varying color & sizes for each data point
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [15,35,40,20,50]
labe = ['Civil', "Electrical","Mechanical",'Chemical','CS']
exp = [0,0,0,0,0.1]

plt.pie(data, labels = labe,explode = exp,  autopct = "%5.2f%%")  #pie chart with CS section exploded out
```

```python
means = [20,35,40,35,27]
stds = [2,3,4,1,2]
ind = np.arange(len(means))
plt.barh(ind,means, color = 'c',label = 'means')
plt.barh(ind+0.5,stds,color = 'k', label = 'stds')  #bar chart with horizontal orientation
plt.legend(loc = 'lower right')
```

```python
x = np.arange(2,20,2)
y = np.log(x)
plt.plot(x,y)
plt.plot(x,y*1.2)
plt.plot(x,y*(-1.2))
plt.xlabel("X-values between 2 and 20, gap is 2 between them")
plt.ylabel("Log of x values")
plt.title("line chart")
```

```python
plt.bar(x,y)
plt.title("Barchart")
```

```python
months = np.arange(1,7)
orders = [200,100,45,50,150,140]
plt.barh(months,orders,color = 'pink')  #horizontal bars graph
plt.xlabel("Orders")
plt.ylabel("Month")
plt.title("Navya's online business sales chart by month")
```
